```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Define the range of p values
p_values = [1,2,3,4,5,6,7,1000]  # Infinity represented as np.inf

# Create a figure to hold the plots
plt.figure(figsize=(10, 8))

# Iterate through different values of p
for p in p_values:
    # Generate theta values
    theta = np.linspace(0, 2 * np.pi, 1000)

    # Calculate the distance for the perimeter using the Minkowski distance formula
    radius = 1


    factor_r = (radius / ((np.abs(np.cos(theta))**p + np.abs(np.sin(theta))**p) **(1/p)))

    # Calculate x and y coordinates from the polar coordinates
    x = factor_r * np.cos(theta)
    y = factor_r * np.sin(theta)

    # Plot the perimeter with the current p value
    plt.plot(x, y, label=f'p={p}')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Perimeter of the Graph Using Minkowski Distance (p Values)')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Equal aspect ratio
plt.show()

```

```python
def plot_Minkowski_distance(p,r):
    theta = np.linspace(0, 2 * np.pi, 1000)
    factor_r = (r / ((np.abs(np.cos(theta))**p + np.abs(np.sin(theta))**p) **(1/p)))
    x = factor_r * np.cos(theta)
    y = factor_r * np.sin(theta)
    plt.figure(figsize=(6, 6))
    plt.plot(x, y, label=f'p={p}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Perimeter of the Graph Using Minkowski Distance with p={p} and radius={r})')
    #plt.legend()
    plt.grid(True)
    plt.axis('equal')  # Equal aspect ratio
    plt.show()
```

```python
plot_Minkowski_distance(1,1)
```
