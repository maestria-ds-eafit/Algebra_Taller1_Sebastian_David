```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of p and radius
p_values = [1, 2, 3, 4, 5, 6 ,7, np.inf]
radius_values = np.arange(0.1, 2.1, 0.1)

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Loop through each value of p
for p in p_values:
    # Calculate the Minkowski distance for each radius value
    distances = [((0, 0), (r, 0)) for r in radius_values]
    minkowski_distances = [((x1 - x2)**p + (y1 - y2)**p)**(1/p) for (x1, y1), (x2, y2) in distances]
    
    # Plot the Minkowski distances
    ax.plot(radius_values, minkowski_distances, label=f'p={p}')

# Add labels and legend
ax.set_xlabel('Radius')
ax.set_ylabel('Minkowski Distance')
ax.set_title('Minkowski Distance Limits with Center (0,0)')
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de p y radios
p_values = [1, 2, 3, 4, 5, 6, 7, 8]  # Puedes ajustar esto según tus necesidades
radios = np.arange(0.1, 2.1, 0.1)

# Crear un conjunto de ángulos
theta = np.linspace(0, 2*np.pi, 100)

# Inicializar la figura
plt.figure(figsize=(8, 8))

# Loop a través de cada valor de p
for p in p_values:
    # Calcular las coordenadas x e y de la circunferencia usando la distancia Minkowski
    x = radios * np.cos(theta)**p
    y = radios * np.sin(theta)**p
    
    # Trazar la circunferencia para el valor de p actual
    plt.plot(x, y, label=f'p={p}')

# Establecer límites y etiquetas de los ejes
#plt.xlim(-2.5, 2.5)
#plt.ylim(-2.5, 2.5)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Título y leyenda
plt.title('Circunferencia usando Distancia Minkowski')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.gca().set_aspect('equal')  # Hacer que los ejes tengan la misma escala
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de p y radios
p_values = [1, 2, 3, 4, 5, 6, 7, 8]  # Puedes ajustar esto según tus necesidades
radios = np.arange(0.1, 2.1, 0.1)

# Crear un conjunto de ángulos
theta = np.linspace(0, 2 * np.pi, 100)

# Inicializar la figura
plt.figure(figsize=(8, 8))

# Loop a través de cada valor de p
for p in p_values:
    # Calcular las coordenadas x e y de la circunferencia usando la distancia Minkowski
    x = radios * np.cos(theta)
    y = radios * np.sin(theta)
    
    # Elevar las coordenadas al orden p
    x_p = (x ** p).reshape(-1, 1)
    y_p = (y ** p).reshape(-1, 1)
    
    # Trazar la circunferencia para el valor de p actual
    plt.plot(x_p, y_p, label=f'p={p}')

# Establecer límites y etiquetas de los ejes
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Título y leyenda
plt.title('Circunferencia usando Distancia Minkowski')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.gca().set_aspect('equal')  # Hacer que los ejes tengan la misma escala
plt.show()
```
